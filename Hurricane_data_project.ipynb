```python
# names of hurricanes
names = ['Cuba I', 'San Felipe II Okeechobee', 'Bahamas', 'Cuba II', 'CubaBrownsville', 'Tampico', 'Labor Day', 'New England', 'Carol', 'Janet', 'Carla', 'Hattie', 'Beulah', 'Camille', 'Edith', 'Anita', 'David', 'Allen', 'Gilbert', 'Hugo', 'Andrew', 'Mitch', 'Isabel', 'Ivan', 'Emily', 'Katrina', 'Rita', 'Wilma', 'Dean', 'Felix', 'Matthew', 'Irma', 'Maria', 'Michael']

# months of hurricanes
months = ['October', 'September', 'September', 'November', 'August', 'September', 'September', 'September', 'September', 'September', 'September', 'October', 'September', 'August', 'September', 'September', 'August', 'August', 'September', 'September', 'August', 'October', 'September', 'September', 'July', 'August', 'September', 'October', 'August', 'September', 'October', 'September', 'September', 'October']

# years of hurricanes
years = [1924, 1928, 1932, 1932, 1933, 1933, 1935, 1938, 1953, 1955, 1961, 1961, 1967, 1969, 1971, 1977, 1979, 1980, 1988, 1989, 1992, 1998, 2003, 2004, 2005, 2005, 2005, 2005, 2007, 2007, 2016, 2017, 2017, 2018]

# maximum sustained winds (mph) of hurricanes
max_sustained_winds = [165, 160, 160, 175, 160, 160, 185, 160, 160, 175, 175, 160, 160, 175, 160, 175, 175, 190, 185, 160, 175, 180, 165, 165, 160, 175, 180, 185, 175, 175, 165, 180, 175, 160]

# areas affected by each hurricane
areas_affected = [['Central America', 'Mexico', 'Cuba', 'Florida', 'The Bahamas'], ['Lesser Antilles', 'The Bahamas', 'United States East Coast', 'Atlantic Canada'], ['The Bahamas', 'Northeastern United States'], ['Lesser Antilles', 'Jamaica', 'Cayman Islands', 'Cuba', 'The Bahamas', 'Bermuda'], ['The Bahamas', 'Cuba', 'Florida', 'Texas', 'Tamaulipas'], ['Jamaica', 'Yucatn Peninsula'], ['The Bahamas', 'Florida', 'Georgia', 'The Carolinas', 'Virginia'], ['Southeastern United States', 'Northeastern United States', 'Southwestern Quebec'], ['Bermuda', 'New England', 'Atlantic Canada'], ['Lesser Antilles', 'Central America'], ['Texas', 'Louisiana', 'Midwestern United States'], ['Central America'], ['The Caribbean', 'Mexico', 'Texas'], ['Cuba', 'United States Gulf Coast'], ['The Caribbean', 'Central America', 'Mexico', 'United States Gulf Coast'], ['Mexico'], ['The Caribbean', 'United States East coast'], ['The Caribbean', 'Yucatn Peninsula', 'Mexico', 'South Texas'], ['Jamaica', 'Venezuela', 'Central America', 'Hispaniola', 'Mexico'], ['The Caribbean', 'United States East Coast'], ['The Bahamas', 'Florida', 'United States Gulf Coast'], ['Central America', 'Yucatn Peninsula', 'South Florida'], ['Greater Antilles', 'Bahamas', 'Eastern United States', 'Ontario'], ['The Caribbean', 'Venezuela', 'United States Gulf Coast'], ['Windward Islands', 'Jamaica', 'Mexico', 'Texas'], ['Bahamas', 'United States Gulf Coast'], ['Cuba', 'United States Gulf Coast'], ['Greater Antilles', 'Central America', 'Florida'], ['The Caribbean', 'Central America'], ['Nicaragua', 'Honduras'], ['Antilles', 'Venezuela', 'Colombia', 'United States East Coast', 'Atlantic Canada'], ['Cape Verde', 'The Caribbean', 'British Virgin Islands', 'U.S. Virgin Islands', 'Cuba', 'Florida'], ['Lesser Antilles', 'Virgin Islands', 'Puerto Rico', 'Dominican Republic', 'Turks and Caicos Islands'], ['Central America', 'United States Gulf Coast (especially Florida Panhandle)']]

# damages (USD($)) of hurricanes
damages = ['Damages not recorded', '100M', 'Damages not recorded', '40M', '27.9M', '5M', 'Damages not recorded', '306M', '2M', '65.8M', '326M', '60.3M', '208M', '1.42B', '25.4M', 'Damages not recorded', '1.54B', '1.24B', '7.1B', '10B', '26.5B', '6.2B', '5.37B', '23.3B', '1.01B', '125B', '12B', '29.4B', '1.76B', '720M', '15.1B', '64.8B', '91.6B', '25.1B']

# deaths for each hurricane
deaths = [90,4000,16,3103,179,184,408,682,5,1023,43,319,688,259,37,11,2068,269,318,107,65,19325,51,124,17,1836,125,87,45,133,603,138,3057,74]

# 1
def convert_to_float(data): 
    """
    Returns a new list where recorded values are converted to floats
    and missing values are retained as 'Damages not recorded'.
    """
    updated = []

    for value in data:
        if value == "Damages not recorded":
            updated.append("Damages not recorded")
        elif value[-1] == "M":
            updated.append(float(value[:-1]) * conversion["M"])
        elif value[-1] == "B":
            updated.append(float(value[:-1]) * conversion["B"])
        else:
            updated.append(float(value))

    return updated

# Update Recorded Damages
conversion = {"M": 1000000,
              "B": 1000000000}

# test function by updating damages
damages = convert_to_float(damages)

# 2 
# Create a Table
# Create and view the hurricanes dictionary

def combine_by_index(*lists):
    """
    Combines multiple lists into a dictionary where each key is the index
    and the value is a list of elements from each input list at that index.
    """
    return {
        i: list(values)
        for i, values in enumerate(zip(*lists))
    }

combined_data = combine_by_index(names, months, years, max_sustained_winds, areas_affected, damages, deaths)
#print(combined_data)

# 3
# Organizing by Year
def hurricanes_by_year(hurricanes):
    """
    Converts the hurricanes dictionary into a new dictionary
    where keys are years and values are lists of hurricane dictionaries.
    """
    hurricanes_year = {}

    for data in hurricanes.values():
        name, month, year, wind, areas, damage, deaths = data

        hurricane_info = {
            "name": name,
            "month": month,
            "wind": wind,
            "areas_affected": areas,
            "damage": damage,
            "deaths": deaths
        }

        if year not in hurricanes_year:
            hurricanes_year[year] = []

        hurricanes_year[year].append(hurricane_info)

    return hurricanes_year
year_organisation = hurricanes_by_year(combined_data)
# create a new dictionary of hurricanes with year and key

#print(year_organisation)
# 4
# Counting Damaged Areas
def count_affected_areas(hurricanes):
    """
    Counts how often each area is affected by hurricanes.
    Returns a dictionary with areas as keys and counts as values.
    """
    area_counts = {}

    for data in hurricanes.values():
        areas = data[4]  # areas_affected is at index 4

        for area in areas:
            if area not in area_counts:
                area_counts[area] = 0
            area_counts[area] += 1

    return area_counts

# create dictionary of areas to store the number of hurricanes involved in

area_counts = count_affected_areas(combined_data)
#print(area_counts)

# 5 
# Calculating Maximum Hurricane Count
def most_affected_area(data):
    """
    Finds the area affected by the most hurricanes.
    Accepts either:
    - a dictionary of {area: count}, or
    - the combined hurricane data dictionary
    Returns (area, count).
    """
    # If values are lists, assume this is combined_data
    if isinstance(next(iter(data.values())), list):
        area_counts = {}

        for hurricane in data.values():
            areas = hurricane[4]  # areas_affected

            for area in areas:
                area_counts[area] = area_counts.get(area, 0) + 1
    else:
        area_counts = data

    max_area = None
    max_count = 0

    for area, count in area_counts.items():
        if count > max_count:
            max_area = area
            max_count = count

    return max_area, max_count



# find most frequently affected area and the number of hurricanes involved in

max_area, max_count = most_affected_area(combined_data)
#print(max_area, max_count)
# 6
# Calculating the Deadliest Hurricane
def deadliest_hurricane(data):
    """
    Finds the hurricane with the highest number of deaths.
    Accepts combined hurricane data dictionary.
    Returns (name, deaths).
    """
    max_deaths = 0
    deadliest = None

    for hurricane in data.values():
        name = hurricane[0]
        deaths = hurricane[6]

        if deaths > max_deaths:
            max_deaths = deaths
            deadliest = name

    return deadliest, max_deaths

# find highest mortality hurricane and the number of deaths
deadliest, max_deaths = deadliest_hurricane(combined_data)

#print(deadliest)
# 7
# Rating Hurricanes by Mortality
def hurricanes_mortality(hurricanes, mortality_scale):
    """
    Rates hurricanes based on the number of deaths.
    
    Parameters:
    - hurricanes: combined hurricane data dictionary
    - mortality_scale: dictionary {rating: upper bound of deaths}
    
    Returns:
    - Dictionary where keys are ratings, and values are lists of hurricane names.
    """
    # Initialize dictionary with empty lists for each rating
    rated = {rating: [] for rating in mortality_scale}

    for hurricane in hurricanes.values():
        name = hurricane[0]
        deaths = hurricane[6]

        # Find the rating this hurricane belongs to
        for rating, upper_bound in sorted(mortality_scale.items()):
            if deaths <= upper_bound:
                rated[rating].append(name)
                break

    return rated

mortality_scale = {0: 0,
                   1: 100,
                   2: 500,
                   3: 1000,
                   4: 10000}

rated = hurricanes_mortality(combined_data, mortality_scale)
#print(rated)

# categorize hurricanes in new dictionary with mortality severity as key


# 8 Calculating Hurricane Maximum Damage
def most_damaging_hurricane(hurricanes):
    """
    Finds the hurricane that caused the greatest damage.
    Returns a tuple: (hurricane_name, damage_amount)
    """
    max_damage = 0
    costliest = None

    for hurricane in hurricanes.values():
        name = hurricane[0]
        damage = hurricane[5]

        # Skip if damage is 'Damages not recorded'
        if isinstance(damage, str):
            continue

        if damage > max_damage:
            max_damage = damage
            costliest = name

    return costliest, max_damage
costliest, max_damage = most_damaging_hurricane(combined_data)

# find highest damage inducing hurricane and its total cost
print(costliest)

# 9

def rate_hurricanes_by_damage(hurricanes, damage_scale):
    """
    Rates hurricanes based on damage amounts.

    Parameters:
    - hurricanes: combined hurricane data dictionary
    - damage_scale: dictionary {rating: upper bound of damage}

    Returns:
    - Dictionary where keys are ratings, values are lists of hurricane names.
    """
    # Initialize dictionary with empty lists for each rating
    rated = {rating: [] for rating in damage_scale}

    for hurricane in hurricanes.values():
        name = hurricane[0]
        damage = hurricane[5]

        # Skip if damage is not a number
        if isinstance(damage, str):
            continue

        # Find which rating this hurricane belongs to
        for rating, upper_bound in sorted(damage_scale.items()):
            if damage <= upper_bound:
                rated[rating].append(name)
                break

    return rated

# Rating Hurricanes by Damage
damage_scale = {0: 0,
                1: 100000000,
                2: 1000000000,
                3: 10000000000,
                4: 50000000000}
rated = rate_hurricanes_by_damage(combined_data, damage_scale)
# categorize hurricanes in new dictionary with damage severity as key
```
